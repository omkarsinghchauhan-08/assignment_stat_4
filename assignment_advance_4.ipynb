```python
#Ques1
#ans-- Both t-tests and z-tests are used to make inferences about population parameters based on sample data. They are both hypothesis tests that help to determine whether there is a significant difference between the means of two groups or between the mean of a sample and a known population mean. However, they differ in their assumptions and when they should be used.

# 1.>  T-Test:
#          A t-test is used when the sample size is small, and the population standard deviation is unknown. It is based on a t-distribution, which has fatter tails compared to the standard normal (z) distribution. The t-distribution takes into account the increased uncertainty due to smaller sample sizes.

# 2.>  Z-Test:
#          A z-test is used when the sample size is large (typically n > 30) or when the population standard deviation is known. It is based on the standard normal (z) distribution, which is a special case of the t-distribution with infinite degrees of freedom.

# Example scenarios:

#    T-Test:
#          Let's say you want to compare the mean test scores of two groups of students (Group A and Group B) to determine if there is a statistically significant difference in their performance. If the sample sizes of both groups are relatively small (e.g., less than 30) and the population standard deviation is unknown, you would use a t-test.

#    Z-Test:
#          Imagine you want to compare the average income of a sample of employees at a company to the average income of employees in the entire industry. If you have a large enough sample size (e.g., more than 30) and you know the population standard deviation (rare, but sometimes available from previous data or studies), then you can use a z-test to assess whether the sample's mean income significantly differs from the industry average.

#> In summary, use a t-test when the sample size is small and the population standard deviation is unknown, and use a z-test when the sample size is large or when the population standard deviation is known.




```

```python
# Ques2
#ans-- one-tailed and two-tailed tests are types of hypothesis tests used to assess the significance of an observed effect or difference in a sample data. The key difference between them lies in the directionality of the alternative hypothesis.

#   One-Tailed Test:
#        In a one-tailed test, the alternative hypothesis makes a specific directional prediction about the population parameter. It tests for the possibility of a significant effect or difference in only one direction (either an increase or a decrease), and it ignores the possibility of an effect in the opposite direction.

#   Example of a one-tailed test:
#  Null hypothesis (H0): The new drug has no effect on improving memory.
#  Alternative hypothesis (Ha): The new drug improves memory performance.

#  Here, we are only interested in testing if the new drug improves memory (one direction).

#  two-Tailed Test:
#     In a two-tailed test, the alternative hypothesis does not specify a particular direction for the effect or difference. It tests for the possibility of a significant effect or difference in either direction (increase or decrease).

#   Example of a two-tailed test:
#     Null hypothesis (H0): There is no difference in test scores between Group A and Group B.
#     Alternative hypothesis (Ha): There is a difference in test scores between Group A and Group B.

#   In this case, we are open to the possibility that there could be a difference in either direction (Group A scoring higher or Group B scoring higher).
```

```python
# Ques 3
# ans -- TYPE 1 ERROR-- We Reject the Null Hypothesis  when in Reality is True.
# example of type 1 Error - Let's say a pharmaceutical company is testing a new drug's effectiveness in reducing headaches. The null hypothesis (H0) is that the drug has no effect on headaches. The alternative hypothesis (Ha) is that the drug reduces headaches.

#  If the company conducts a clinical trial and mistakenly concludes that the drug reduces headaches (rejecting H0), but in reality, the drug has no effect on headaches, it is a Type 1 error. They falsely claim that the drug is effective when it is not.

# Type 2 Error-- We Retain the Null Hypothesis , when in reality it is false.
# example of type 2 Error - Continuing with the pharmaceutical company example, suppose the null hypothesis (H0) is that the drug has no effect on headaches, and the alternative hypothesis (Ha) is that the drug reduces headaches.

#  If the company conducts a clinical trial, but they fail to find enough evidence to conclude that the drug reduces headaches (not rejecting H0), even though the drug does reduce headaches in reality, it is a Type 2 error. They falsely claim that the drug is ineffective when it actually does work.
```

```python
# Ques 6 
# ans -- Sample Problem:
#Suppose a factory produces two types of products: Product A and Product B. Historically, 20% of the factory's output is Product A, and the rest (80%) is Product B. The factory has a quality control system that correctly identifies 90% of Product A as defective and 95% of Product B as non-defective. Now, a randomly selected product is tested, and it is found to be defective. What is the probability that the product is actually Product A?

# Solution using Bayes's Theorem:

#Let's define the events:

# A:Product A
# B: Product B
# D: Defective product

# We want to find 

# we want to find P(A∣D), the probability that the product is Product A given that it is defective.
#  P(A)=0.20
#  P(B)= 0.80
# P(D|A)=0.90
# P(D|B)=0.05

#  Calculate the probability of the evidence (defective product):
# P(D)=P(D|A)*P(A) + P(D|B)*P(B)
#      =0.90*0.20 + 0.05*0.80
#      =0.18 +0.04
# P(D) =0.22  


# Apply Bayes's Theorem:
# P(A|D)= P(D|A).P(A)/P(D)
#      =0.18/0.22
#      =0.8182



#  Interpret the result:
#   The probability that the product is Product A given that it is defective (based on the quality control system's results) is approximately 0.8182 or 81.82%.

# So, if a randomly selected product is found to be defective, there is an 81.82% chance that it is Product A.
```
